```python
import random
import numpy as np
```

```python
#1a
z=[]
def distribution(sig, x0):
    theta = 2*np.pi*random.random()
    r = (-2*np.log(1 - random.random()))**(1/2)
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    
    return x * sig + x0

#1b
for i in range(0,1001):
    z.append(distribution(-1, 2))
z
```

```python
#Code from powerpoint

NT1 = 1000
NPb = 0 
tau = 3.053*60
h = 1.0
p = 1 - 2**(-h/tau)
tmax = 1000

tpoints = np.arange(0.0, tmax, h)
TLpoints = []
PBpoints = []

for t in tpoints:
    TLpoints.append(NT1)
    PBpoints.append(NPb)
    
    decay = 0
    for i in range(NT1):
        if random.random() < p:
            decay += 1
    NT1 -= decay
    NPb += decay

    
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(tpoints, TLpoints, s=5, c='b', alpha=0.3)
ax.scatter(tpoints, PBpoints, s=5, c='r', alpha=0.3)
ax.set_xlabel("$Time$ $(s)$", size=20)
ax.xaxis.set_tick_params(labelsize=20)
ax.yaxis.set_tick_params(labelsize=20)
```

```python
#2
BI213 = 10000
TI1 = 0
NPb = 0 
Bi209 = 0
tau1 = 46*60
tau2 = 2.2*60
tau3 = 3.3*60
h = 1.0
p = 1 - (2**(-h/tau1))
p_TI = 1 - (2**(-h/tau2))
p_NPb = 1 - (2**(-h/tau3))
tmax = 20000

tpoints = np.arange(0.0, tmax, h)
BI213points = []
PBpoints = []
TIpoints = []
Bi209points = []


for t in tpoints:
    BI213points.append(BI213)
    PBpoints.append(NPb)
    TIpoints.append(TI1)
    Bi209points.append(Bi209)


    decay = 0
  
    for i in range(NPb):
        if random.random() < p_NPb:
            decay += 1
        NPb -= decay
        Bi209 += decay
    

    for i in range(TI1):
        if random.random() < p_TI:
            decay += 1
        TI1 -= decay
        NPb += decay
        
        
    for i in range(BI213):
        if random.random() < p:
            decay += 1
    BI213 -= decay
    
    if random.random() <= 0.9791:
        NPb += decay
        BI213 -= decay
    else:
        TI1 += decay
        BI213 -= decay

    
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)
ax.scatter(tpoints, BI213points, s=5, c='y', alpha=0.3)
ax.scatter(tpoints, PBpoints, s=5, c='r', alpha=0.3)
ax.scatter(tpoints, TIpoints, s=5, c='m', alpha=0.3)
ax.scatter(tpoints, Bi209points, s=5, c='c', alpha=0.3)
ax.set_xlabel("$Time$ $(s)$", size=20)
ax.set_ylabel("$Atoms$", size=20)
ax.xaxis.set_tick_params(labelsize=20)
ax.yaxis.set_tick_params(labelsize=20)



#Yellow is the Bi 213 over time
#Red is the Pb over time
#magenta is the Ti over time
#Cyan is the Bi 209 over time
```
